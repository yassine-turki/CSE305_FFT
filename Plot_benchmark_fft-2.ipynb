```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python


# Read the CSV file
df = pd.read_csv("execution_times.csv")
df = df.rename(columns=lambda x: x.replace(' threads', '').strip())
df = df.rename(columns=lambda x: x.replace('Sequential', 'Seq.').strip())

# Set the algorithm as index
df.set_index("Algorithm", inplace=True)

# Plot line chart
for algorithm in df.index:
    plt.plot(df.columns, df.loc[algorithm], marker="o", label=algorithm)

# Add labels and title
plt.xlabel("Number of threads")
plt.ylabel("Time (seconds)")
plt.title("Benchmark results for input of 10E4 points")

# Add legend
plt.legend()

# Set y-axis scale to logarithmic
plt.yscale("log")

# Show the plot
plt.show()

```
